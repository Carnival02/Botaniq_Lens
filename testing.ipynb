```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
cnn=tf.keras.models.load_model('C:\\Users\\harsh\\OneDrive\\Desktop\\fruits_Recognition_system\\Plant_deasieas\\trained_model.keras')
```

```python
#performing prediction on single image

import cv2
image_path="C:\\Users\\harsh\\OneDrive\\Desktop\\fruits_Recognition_system\\Plant_deasieas\\train\\orange\\Image_1.png"
img=cv2.imread(image_path)
plt.imshow(img)
plt.title("Test Image")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#testing Model

image=tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64))
input_arr=tf.keras.preprocessing.image.img_to_array(image)
input_arr=np.array([input_arr])
prediction=cnn.predict(input_arr)
```

```python
print(prediction)
print(max(prediction[0]))
```
